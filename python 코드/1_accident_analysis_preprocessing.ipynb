```python
import pandas as pd
import numpy as np
```

## Data load

```python
# 데이터 불러오기
kp20 = pd.read_csv('_1_accident analysis code csv파일//KP2020.csv', encoding='cp949')
kp21 = pd.read_csv('_1_accident analysis code csv파일/KP2021.csv', encoding='cp949')
npa20 = pd.read_csv('_1_accident analysis code csv파일//NPA2020.csv', encoding='cp949')
```

## Data merge

```python
# EVT_CL_CD가 401(교통사고)인 행 추출
kp2020 = kp20[kp20['EVT_CL_CD']==401]
kp2021 = kp21[kp21['EVT_CL_CD']==401]
npa2020 = npa20[npa20['EVT_CL_CD']==401]
```

```python
# kp2020 RECV_CPLT_DM 컬럼 값 변경 (뒤에 0000000 제거)
split1 = kp2020.RECV_CPLT_DM.str.split('.')
a_datetime = split1.str.get(0)
kp2020['RECV_CPLT_DM'] = a_datetime
kp2020
```

```python
# kp2021 RECV_CPLT_DM 컬럼 값 변경 (뒤에 0000000 제거)
split2 = kp2021.RECV_CPLT_DM.str.split('.')
b_datetime = split2.str.get(0)
kp2021['RECV_CPLT_DM'] = b_datetime
kp2021
```

```python
#kp2021 kp2020 데이터 합치기
kp_merged = pd.concat([kp2020,kp2021], axis=0)
kp_merged.reset_index(inplace=True)
kp_merged = kp_merged.drop(['index'], axis=1)
kp_merged
```

```python
#kp_merged RECV_CPLT_DM컬럼으로 RECV_date, RECV_time split 컬럼 생성
split3 = kp_merged.RECV_CPLT_DM.str.split(' ')
a_date = split3.str.get(0)
a_time = split3.str.get(1)
kp_merged['RECV_date'],kp_merged['RECV_time'] = (a_date,a_time)
```

```python
#RECV_date컬럼 year,month,day로 split
date_split = kp_merged.RECV_date.str.split('/')
a_year = date_split.str.get(0)
a_month = date_split.str.get(1)
a_day = date_split.str.get(2)
kp_merged['RECV_year'],kp_merged['RECV_month'],kp_merged['RECV_day'] = (a_year,a_month,a_day)
kp_merged = kp_merged.drop(['RECV_date'], axis=1)
```

```python
#RECV_time컬럼 hour,minute,second로 split
time_split = kp_merged.RECV_time.str.split(':')
a_hour = time_split.str.get(0)
a_minute = time_split.str.get(1)
a_second = time_split.str.get(2)
kp_merged['RECV_hour'],kp_merged['RECV_minute'],kp_merged['RECV_second'] = (a_hour,a_minute,a_second)
kp_merged = kp_merged.drop(['RECV_time'], axis=1)
```

```python
kp_merged
```

```python
npa2020
```

```python
#npa20020 HPPN_OLD_ADDR 컬럼 명칭 변경
npa2020 = npa2020.rename(columns={'HPPN_OLD_ADDR':'HPPN_PNU_ADDR'})
```

```python
#npa2020의 날짜와 시간 컬럼 kp_merged와 형식 맞추기
npa2020['RECV_CPLT_TM'] = npa2020["RECV_CPLT_TM"].astype('str').str.zfill(6)
npa2020['RECV_CPLT_DT'] = npa2020["RECV_CPLT_DT"].astype('str')
npa2020['RECV_CPLT_DT'] = npa2020["RECV_CPLT_DT"].apply(lambda x : x[2:])
npa2020.head()
```

```python
#npa2020 year / month / day / hour / minute / second 컬럼 생성
npa2020['RECV_year'] = npa2020["RECV_CPLT_DT"].apply(lambda x : x[0:2])
npa2020['RECV_month'] = npa2020["RECV_CPLT_DT"].apply(lambda x : x[2:4])
npa2020['RECV_day'] = npa2020["RECV_CPLT_DT"].apply(lambda x : x[4:])
npa2020['RECV_hour'] = npa2020["RECV_CPLT_TM"].apply(lambda x : x[0:2])
npa2020['RECV_minute'] = npa2020["RECV_CPLT_TM"].apply(lambda x : x[2:4])
npa2020['RECV_second'] = npa2020["RECV_CPLT_TM"].apply(lambda x : x[4:])
npa2020.head()
```

```python
#npa2020 RECV_CPLT_DM 컬럼 생성 및 RECV_CPLT_DT / RECV_CPLT_TM 컬럼 제거
npa2020['RECV_year'] = npa2020["RECV_year"].astype('str')
npa2020['RECV_month'] = npa2020["RECV_month"].astype('str')
npa2020['RECV_day'] = npa2020["RECV_day"].astype('str')
npa2020['RECV_hour'] = npa2020["RECV_hour"].astype('str')
npa2020['RECV_minute'] = npa2020["RECV_minute"].astype('str')
npa2020['RECV_second'] = npa2020["RECV_second"].astype('str')
npa2020['RECV_CPLT_DM'] = npa2020['RECV_year']+'/'+npa2020['RECV_month']+'/'+npa2020['RECV_day']+' '+npa2020['RECV_hour']+':'+npa2020['RECV_minute']+':'+npa2020['RECV_second']
npa2020 = npa2020.drop(['RECV_CPLT_DT'], axis=1)
npa2020 = npa2020.drop(['RECV_CPLT_TM'], axis=1)
```

```python
#불필요한 컬럼 RECV_DEPT_NM 제거
kp_merged = kp_merged.drop(['RECV_DEPT_NM'], axis=1)
```

```python
#kp_merged 와 npa2020 데이터 합치기
accident = pd.concat([kp_merged,npa2020], axis=0)
accident
```
