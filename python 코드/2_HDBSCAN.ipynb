```python
import hdbscan
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daejeon = pd.read_csv('_2_HDBSCAN csv파일//daejeon.csv', encoding = 'utf-8', index_col = 0)
sejong = pd.read_csv('_2_HDBSCAN csv파일//sejong.csv', encoding = 'utf-8', index_col = 0)
chungnam = pd.read_csv('_2_HDBSCAN csv파일//chungnam.csv', encoding = 'utf-8', index_col = 0)
display(daejeon)
display(sejong)
display(chungnam)
```

### 대전 

```python
kms_per_radian = 6371.0088
#epsilon /= kms_per_radian

clusterer_dae = hdbscan.HDBSCAN(min_cluster_size =50, min_samples = 50, metric = 'haversine', cluster_selection_method = 'eom')
clusterer_dae.fit(
    np.radians(
        [x for x in zip(daejeon['HPPN_Y'],daejeon['HPPN_X'])]
    )
)
```

```python
np.unique(clusterer_dae.labels_, return_counts = True)
```

```python
daejeon['cluster'] = clusterer_dae.labels_
```

```python
daejeon = daejeon[daejeon['cluster']!='75'].reset_index()
```

```python
daejeon.to_csv('_2_HDBSCAN csv파일//daejeon_eom_50.csv')
```

### 노이즈 제외 추출

```python
d_wn = d_50[d_50['cluster'] != -1]
c_wn = c_40[c_40['cluster'] != -1]
s_wn = s_50[s_50['cluster'] != -1]

display(d_wn)
display(c_wn)
display(s_wn)
```

```python
from datetime import datetime
d_wn['RECV_CPLT_DM'] = pd.to_datetime(d_wn['RECV_CPLT_DM'])
c_wn['RECV_CPLT_DM'] = pd.to_datetime(c_wn['RECV_CPLT_DM'])
s_wn['RECV_CPLT_DM'] = pd.to_datetime(s_wn['RECV_CPLT_DM'])
```
