### 기상 데이터 병합

```python
#연도별 기상데이터 병합
def merge_weather(data20,data21,data22,data23) :
    city_wea = pd.concat([data20,data21,data22], axis=0)
    city_wea = city_wea[['일시', '평균기온(°C)', '일강수량(mm)','평균 풍속(m/s)','평균 상대습도(%)','평균 이슬점온도(°C)']]
    data23 = data23.drop(['지점','지점명'], axis=1)
    city_wea = pd.concat([city_wea,data23], axis=0)
    city_wea = city_wea.rename(columns={'일시':'date'})
    city_wea['일강수량(mm)'] = city_wea['일강수량(mm)'].fillna(0.0)
    return city_wea
```

### cluster별로 뽑아서 캘린더 csv 만들기

```python
def extract_cluster(data,num):
    
    data = data[['RECV_CPLT_DM','cluster']]
    df1 = data[data['cluster']==num]
    df1['accident_num'] =1 
    df2 = df1.groupby([df1['RECV_CPLT_DM'].dt.date]).sum()
    df2 = df2.reset_index()
    df2['date'] = pd.to_datetime(df2['RECV_CPLT_DM'])
    
    df3 = pd.DataFrame()
    df3['date'] =pd.date_range("2020-1-1", "2023-1-31")
    df4 = pd.merge(df3, df2, on='date', how='outer')
    df4['accident_num'] = df4['accident_num'].fillna(0)
    df4 = df4.drop(['RECV_CPLT_DM'], axis = 1)
    return df4
```

### 지역 cluster별 캘린더와 기상 및 날짜 데이터 합치기

```python
import os

def merge_num_wea(city, wea_data):
    path = '_6_accident_prediction_preprocessing//'+str(city)+'_clu//'
    file_list = os.listdir(path)
    for i in file_list:
        df1 = pd.read_csv(path + i)
        clu_num = i.split('.')[0]
        df_merged = pd.merge(df1, wea_data, on='date', how='left')
        df_merged['date'] = pd.to_datetime(df_merged['date'])
        df_merged.set_index('date', inplace=True)
        df_merged['year']=df_merged.index.year
        df_merged['month']=df_merged.index.month 
        df_merged['day']=df_merged.index.day
        df_merged = df_merged.reset_index()
        df_merged['date'] = pd.to_datetime(df_merged['date'])
        df_merged.to_csv('_6_accident_prediction_preprocessing//'+str(city)+'//'+str(city)+'_'+str(clu_num)+".csv",mode="w",encoding='utf-8',index=False)
    return 'Saved Successfully '
```

### 미세먼지 데이터 합치기

```python
# 수치 카테고리화
def dust_categorize(data):
    data['O3']=pd.cut(data['O3'],bins=[0,0.03,0.09,0.15,3],labels=[0,1,2,3],right=True)
    data['PM2.5']=pd.cut(data['PM2.5'],bins=[0,15,35,75,500],labels=[0,1,2,3],right=True) 
    data['PM10']=pd.cut(data['PM10'],bins=[0,30,80,150,500],labels=[0,1,2,3],right=True)
    return 'Saved Successfully'
```

```python
# spline 보간법
def do_spline(data):
    data['평균기온(°C)']= data['평균기온(°C)'].interpolate(option='spline')
    data['평균 상대습도(%)']= data['평균 상대습도(%)'].interpolate(option='spline')
    data['평균 이슬점온도(°C)']= data['평균 이슬점온도(°C)'].interpolate(option='spline')
    data['O3']= data['O3'].interpolate(option='spline')
    data['PM10']= data['PM10'].interpolate(option='spline')
    data['PM2.5']= data['PM2.5'].interpolate(option='spline')
    return 'Saved Successfully'
```

### 추가적인 변수 생성

```python
# 계절 컬럼 생성
def what_season(data) :
    data['season']=pd.cut(data['month'],bins=[0,2,5,8,11,12],labels=['winter','spring','summer','fall','winter'],ordered=False,right=True)
    return "Saved Successfully!"
```

```python
# 요일 컬럼 생성
def day_of_the_week (data) :
    data['date'] = pd.to_datetime(data['date'])
    data['date'].dt.dayofweek
    data['요일'] = data['date'].dt.dayofweek
    data.loc[data['요일']==0, '요일'] = '월요일'
    data.loc[data['요일']==1, '요일'] = '화요일'
    data.loc[data['요일']==2, '요일'] = '수요일'
    data.loc[data['요일']==3, '요일'] = '목요일'
    data.loc[data['요일']==4, '요일'] = '금요일'
    data.loc[data['요일']==5, '요일'] = '토요일'
    data.loc[data['요일']==6, '요일'] = '일요일'
    return 'Saved Successfully!'
```
